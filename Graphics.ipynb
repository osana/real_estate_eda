# Анализ данных по графикам

```python
import re
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import requests
path1 = "./datasets/_data_fixed.csv"
path2 = './datasets/_data.csv'
rent_fixed = pd.read_csv(path1)
rent = pd.read_csv(path2)
rent.dtypes
```

```python
nan = rent.isna().sum()
nan.plot(kind='bar')
plt.title('NaN values')
plt.xlabel('Columns')
plt.show()
```

Зависмость площади квартиры от стоимости аренды в месяц.

```python
sns.scatterplot(data=rent_fixed, x='price_rub_month', y='square_total')
plt.title('The flat square and rent per moth correlation')
plt.xlabel('Price per month, RUB')
plt.ylabel('Square, m2')
plt.show()
```

Стоимость аренды квартиры в зависимости от округа.

```python
sns.barplot(data=rent_fixed, x='price_rub_month', y='Округ')

plt.title('The flat rent depending on the district')
plt.xlabel('Price per month, RUB')
plt.ylabel('District')
plt.show()
```
